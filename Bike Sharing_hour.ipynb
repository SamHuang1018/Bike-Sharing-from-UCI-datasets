```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
plt.style.use('seaborn')
plt.figsize = (10, 6)

import warnings
warnings.filterwarnings("ignore")
```

```python
df1 = pd.read_csv('hour.csv', index_col=1, parse_dates=True)
df1 = df1.drop(columns=['instant', 'casual', 'registered'])
df1.head()
```

```python
for i in ['season', 'mnth', 'hr', 'weekday', 'weathersit']:
    df1[i] = df1[i].astype(str)

df1.info()
```

```python
df1_1 = pd.get_dummies(df1)
df1_1.head()
```

```python
df1_1.columns
```

```python
df1_1.shape
```

```python
df2=df1_1.sample(frac=1, random_state=100)

for i in range(9):
    df2=df2.sample(frac=1, random_state=100)
```

```python
df2.head()
```

```python
N = len(df1)
N1 = int(4 * N / 6)
N2 = int(5 * N / 6)
```

```python
x_train = df2.drop(columns=['cnt']).iloc[:N1]
```

```python
x_train.describe()
```

```python
x_train.shape
```

```python
x_val = df2.drop(columns=['cnt']).iloc[N1:N2]
x_val.describe()
```

```python
x_val.shape
```

```python
x_test = df2.drop(columns=['cnt']).iloc[N2:]
x_test.describe()
```

```python
x_test.shape
```

```python
y_train = df2['cnt'].iloc[:N1]
```

```python
y_train.describe()
```

```python
y_val = df2['cnt'].iloc[N1:N2]
```

```python
y_val.describe()
```

```python
y_test = df2['cnt'].iloc[N2:]
```

```python
y_test.describe()
```

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
sc.fit(x_train)

x_train_std = sc.transform(x_train)
x_val_std = sc.transform(x_val)
x_test_std = sc.transform(x_test)
```

# Linear Regression

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression().fit(x_train_std, y_train)
model.score(x_test_std, y_test)
```

# Ridge

```python
from sklearn.linear_model import Ridge

c1 = []

for i in range(-4, 5):
    model = Ridge(alpha = 10 ** i, random_state=10000)
    model.fit(x_train_std, y_train)
    print('The score for alpha = {} is {:.6f}.'.format(10 ** i, model.score(x_val_std, y_val)))
    c1.append(model.score(x_val_std, y_val))

print()

c2 = np.argmax(c1)
print('The proper alpha is {}.'.format(10 ** (int(c2) - 4)))
```

```python
model =  Ridge(alpha = 10 ** (int(c2) - 4), random_state=100)
model.fit(x_train_std, y_train)

model.score(x_test_std, y_test)   
```

# Lasso

```python
from sklearn.linear_model import Lasso

c1 = []

for i in range(-4, 5):
    model =  Lasso(alpha = 30 ** i, random_state=10000)
    model.fit(x_train_std, y_train)
    print('The score for alpha = {} is {:.6f}.'.format(10 ** i, model.score(x_val_std, y_val)))
    c1.append(model.score(x_val_std, y_val))

print()

c2 = np.argmax(c1)
print('The proper alpha is {}.'.format(10 ** (int(c2) - 4)))
```

```python
model =  Lasso(alpha = 10 ** (int(c2) - 4), random_state=100)
model.fit(x_train_std, y_train)

model.score(x_test_std, y_test)   
```

# Elastic Net 

```python
from sklearn.linear_model import ElasticNet

c1 = np.zeros((9, 9))

for L1 in range(-4, 5):
    print('L1 =', 10 ** L1)  
    
    for L2 in range(-4, 5):
        alpha = 150 ** L1 + 150 ** L2
        l1_ratio = 150 ** L1 / alpha
        model =  ElasticNet(alpha = alpha, l1_ratio = l1_ratio, random_state=1000)
        model.fit(x_train_std, y_train)
        print('The score for L2 = {} is {:.6f}.'.format(10 ** L2, model.score(x_val_std, y_val)))
        c1[L1 + 4, L2 + 4] = model.score(x_val_std, y_val)
        
    print()

c2 = np.unravel_index(np.argmax(c1), c1.shape)
print(c2)

print('The proper L1 and L2 are {} and {}.'.format(10 ** (int(c2[0]) - 4), 10 ** (int(c2[1]) - 4)))
```

```python
alpha = 10 ** (int(c2[0]) - 4) + 10 ** (int(c2[1]) - 4)
l1_ratio = 10 ** (int(c2[0]) - 4) / alpha

model = ElasticNet(alpha = alpha, l1_ratio = l1_ratio, random_state=100)
model.fit(x_train_std, y_train)

model.score(x_test_std, y_test)  
```

## Support Vector Regression (So Slow! Guess C = 1000 and epsilon = 2.0)

```python
from sklearn.svm import SVR

c1 = np.zeros((3, 20))

for i in range(2, 5):
    print('C =', 10 ** i)
    
    for j in range(1, 21):
        model = SVR(C = 10 ** i, epsilon = j / 10)
        model.fit(x_train_std, y_train)
        print('The score for C = {} and epsilon = {} is {:.6f}.'.format(10 ** i, j / 10, model.score(x_val_std, y_val)))
        c1[i - 2, j - 1] = model.score(x_val_std, y_val)
    
    print()

c2 = np.unravel_index(np.argmax(c1), c1.shape)
print(c2)

print('The proper C and epsilon are {} and {}.'.format(10 ** (int(c2[0]) + 2), (int(c2[1]) + 1) / 10))
```

## Neutral Network - Tensorflow Keras

```python
from tensorflow import keras
from tensorflow.keras import layers
```

```python
model = keras.Sequential([
    layers.Dense(128, kernel_initializer='he_normal',
                 activation='relu'),
    layers.Dropout(0.2),
    layers.Dense(256, kernel_initializer='he_normal',
                 activation='relu'),
    layers.Dropout(0.2), 
    layers.Dense(512, kernel_initializer='he_normal',
                 activation='relu'),
    layers.Dropout(0.1),
    layers.Dense(1)
])

model.compile(optimizer='adam',
              loss='mse',
              metrics=['mae'])


callbacks_list = [                                   
    keras.callbacks.EarlyStopping(                   
        monitor='val_loss',                          
        patience=50,                                
    ),

    keras.callbacks.ReduceLROnPlateau(
        monitor='val_loss', 
        factor=0.5,
        patience=20, 
        min_lr=0.00001,
    ),

    keras.callbacks.ModelCheckpoint(                 
        filepath='Bike Sharing_hour.h5',               
        monitor='val_loss',                          
        save_best_only=True,                         
    )  
]


history = model.fit(x_train_std,
                    y_train,
                    epochs=10000,
                    callbacks=callbacks_list,
                    batch_size=512,
                    validation_data=(x_val_std, y_val))
```

```python
pd.DataFrame(history.history)
```

```python
loss = history.history['loss'][20:]
val_loss = history.history['val_loss'][20:]

epochs = range(21, len(loss) + 21)

plt.plot(epochs, loss, 'bo', label='Training loss')
plt.plot(epochs, val_loss, 'b', label='Validation loss')

plt.title('Training and validation loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```
